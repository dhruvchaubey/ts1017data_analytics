# Data Aggregation & Visualization

```python
# Aggregating data to identify key patterns

# 1. Flight delay analysis: Number and percentage of delayed flights
delayed_flights = df['Flight Delayed?'].value_counts(normalize=True) * 100

# 2. Lounge usage analysis: Lounge used vs. not used
lounge_usage = df['Lounge Used?'].value_counts(normalize=True) * 100

# 3. Snack analysis: Planned vs. additional snacks
planned_snack = df['Planned Snack?'].value_counts(normalize=True) * 100
additional_snack = df['Additional Snack?'].value_counts(normalize=True) * 100

# 4. Baggage analysis: Average number of included and total checked bags
avg_included_bags = df['# of Included Checked Bags'].mean()
avg_total_bags = df['Total # of Checked Bags'].mean()

# Summarizing key findings
summary = {
    'Delayed Flights (%)': delayed_flights,
    'Lounge Usage (%)': lounge_usage,
    'Planned Snack (%)': planned_snack,
    'Additional Snack (%)': additional_snack,
    'Average Included Bags': avg_included_bags,
    'Average Total Bags': avg_total_bags
}

summary
```

```python
# Aggregate the number of delayed flights by departing airport
delays_by_airport = df.groupby('Departing Airport')['Flight Delayed?'].mean()

# Plotting
plt.figure(figsize=(10, 6))
sns.barplot(x=delays_by_airport.index, y=delays_by_airport.values)
plt.title('Average Flight Delay by Departing Airport')
plt.xlabel('Departing Airport')
plt.ylabel('Percentage of Delayed Flights')
plt.xticks(rotation=90)
plt.show()
```

```python
# Aggregate lounge usage by inquiry type
lounge_by_inquiry = df.groupby('Inquiry Type')['Lounge Used?'].mean()

# Plotting
plt.figure(figsize=(10, 6))
sns.barplot(x=lounge_by_inquiry.index, y=lounge_by_inquiry.values, palette="viridis")
plt.title('Lounge Usage by Inquiry Type')
plt.xlabel('Inquiry Type')
plt.ylabel('Percentage of Lounge Usage')
plt.xticks(rotation=45)
plt.show()
```

```python
# Aggregate average number of checked bags by inquiry type
baggage_by_inquiry = df.groupby('Inquiry Type')['Total # of Checked Bags'].mean()

# Plotting
plt.figure(figsize=(10, 6))
sns.barplot(x=baggage_by_inquiry.index, y=baggage_by_inquiry.values, palette="Set2")
plt.title('Average Number of Checked Bags by Inquiry Type')
plt.xlabel('Inquiry Type')
plt.ylabel('Average Checked Bags')
plt.xticks(rotation=45)
plt.show()
```

```python
# Aggregate planned and additional snacks by departing airport
snacks_by_airport = df.groupby('Departing Airport')[['Planned Snack?', 'Additional Snack?']].mean()

# Plotting
snacks_by_airport.plot(kind='bar', stacked=True, figsize=(12, 6), colormap='Paired')
plt.title('Snack Preferences by Departing Airport')
plt.ylabel('Percentage')
plt.xlabel('Departing Airport')
plt.xticks(rotation=90)
plt.show()
```
