Fixing the time zone values from the columns Departure time zone and Arrival time zones

```python
# Create a mapping of regex patterns to their corresponding timezone values
timezone_mapping = {
    r'^UT.*$': 'UTC',  # Any value starting with 'UT' to 'UTC'
    r'^MS.*$': 'MST',  # Any value starting with 'MS' to 'MST'
    r'^ES.*$': 'EST',  # Any value starting with 'ES' to 'EST'
    r'^PS.*$': 'PST',  # Any value starting with 'PS' to 'PST'
    r'^CS.*$': 'CST',  # Any value starting with 'CS' to 'CST'
}

# Replace using the mapping
for pattern, replacement in timezone_mapping.items():
    df['Arrival Time Zone'] = df['Arrival Time Zone'].replace(to_replace=pattern, value=replacement, regex=True)

# Show the resulting DataFrame
#print(df['Departure Time Zone'])

timezone_mapping = {
    r'^UT.*$': 'UTC',  # Any value starting with 'UT' to 'UTC'
    r'^MS.*$': 'MST',  # Any value starting with 'MS' to 'MST'
    r'^ES.*$': 'EST',  # Any value starting with 'ES' to 'EST'
    r'^PS.*$': 'PST',  # Any value starting with 'PS' to 'PST'
    r'^CS.*$': 'CST',  # Any value starting with 'CS' to 'CST'
}

# Replace using the mapping
for pattern, replacement in timezone_mapping.items():
    df['Departure Time Zone'] = df['Departure Time Zone'].replace(to_replace=pattern, value=replacement, regex=True)

# Show the resulting DataFrame
#print(df['Departure Time Zone'])

df.head()
```
